```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
fpath = glob('./female/*.jpg') #this will return all image path in a list
mpath = glob('./male/*.jpg')
```

```python
print('The num of image in Famale folder: ', len(fpath))
print('The num of image in male folder: ', len(mpath))
```

```python
# read the image  and convert to RGB
# apply hear cascade classfier
# crop face
# save image
```

```python
for i in range(len(fpath)):
    try:
        img = cv2.imread(fpath[i]) #read in BGR format
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)# read the image  and convert to RGB

        haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml') # apply hear cascade classfier
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
        faces_list = haar.detectMultiScale(gray, 1.3, 5)
        # print(faces_list)


        for x,y,w,h in faces_list:
            cv2.rectangle(img_rgb, (x,y),(x+w,y+h),(0,255,0),2)
            # crop face
            roi = img[y:y+h, x:x+w]
            # save image
            cv2.imwrite(f'./crop_data/female/female_{i}.jpg', roi)
            print('Image process success processed')
    except:
        print('Unable to process image')
```
